```python
import pandas as pd
```

```python
df_urgencias.columns
```

```python
df_urgencias_cdmx = df_urgencias[df_urgencias['CLUES'].str.startswith('DF')]
df_urgencias_cdmx
```

```python
columnas_seleccionadas = ['CLUES','EDAD','SEXO','ENTRESIDENCIA', 'DERHAB','TIPOURGENCIA','MOTATE','TIPOCAMA','ENVIADOA','fechaingreso','HORASESTANCIA','hora_ingreso']
df_filtrado = df_urgencias_cdmx[columnas_seleccionadas]
```
